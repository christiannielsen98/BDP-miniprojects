```python
import pandas as pd

KDD_major_mislabel_clean_raw = pd.read_csv('../data/KDD_uniform/raw/mislabel_clean_raw.csv')
KDD_major_raw = pd.read_csv('../data/KDD_uniform/raw/raw.csv')
```

```python
concat_dfs = pd.concat([KDD_major_raw, KDD_major_mislabel_clean_raw]).drop_duplicates(keep=False)

error_cols = list(concat_dfs.columns[concat_dfs.nunique() <= 2])
error_index = list(concat_dfs.index.drop_duplicates())
```

```python
KDD_wrong = KDD_major_raw.iloc[error_index][error_cols]
KDD_correct = KDD_major_mislabel_clean_raw.iloc[error_index][error_cols]
error_subset = KDD_wrong.join(KDD_correct, lsuffix='_wrong', rsuffix='_correct')
```

```python
print(len(KDD_wrong), len(KDD_correct), len(error_subset))
```

```python
error_subset.drop_duplicates().sort_index()
error_subset = error_subset[sorted(error_subset.columns.to_list())]
```

```python
merged_error_cols = []
for col in error_cols:
    merged_error_cols.append(col + '_wrong')
    merged_error_cols.append(col + '_correct')
merged_error_cols = sorted(merged_error_cols)
```
