```python
movie_raw = pd.read_csv('../Data/Movie/inconsistency/dirty_train.csv')
movie_clean = pd.read_csv('../Data/Movie/inconsistency/clean_train.csv')
```

```python
movie_raw
```

```python
movie_raw_and_clean = movie_raw.merge(movie_clean, how='left', left_index=True, right_index=True, suffixes=('_raw', '_clean'))
```

```python
movie_raw_and_clean[movie_raw_and_clean['title_raw'].str.contains('I Am Love')]
```

```python
import json
movie_lang_questions_json = [{"question": f"What language is spoken in the movie {i}?", "answer": f"{j}", "type": "bridge"} for i, j in movie_raw_and_clean[['title_raw', 'language_clean']].values]

with open("movie_lang_questions_json.json", "w") as outfile:
    json.dump(movie_lang_questions_json, outfile)
```
